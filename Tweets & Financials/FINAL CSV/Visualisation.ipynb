```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
```

```python
closing = pd.read_csv("tesla_closing.csv")
```

```python
closing = closing.set_index('Date')
```

```python
closing.head()
```

```python
closing_plot = closing["Adj Close"].plot(linewidth=1, figsize=(15,10));
closing_plot.set_xlabel("Date", fontsize=20)
closing_plot.set_ylabel("Adj Closing Share Price", fontsize=20)
closing_plot.set_title("Tesla Stock Price, July 15 - June 20", fontsize=20)
closing_plot.savefig('share_price.pdf')
```

```python
dataset = pd.read_csv("combined.csv")
```

```python
dataset = dataset.set_index('Date')
```

```python
dataset.head()
```

```python
sentiment = dataset.drop(['Daily Change', 'Direction', 'Signal'], axis=1)
```

```python
sentiment_plot = sentiment.plot(linewidth=1, figsize=(15,10));
sentiment_plot.set_xlabel("Date", fontsize=20)
sentiment_plot.set_ylabel("Sentiment", fontsize=20)
sentiment_plot.set_title("Sentiment Analysis, 07/15 - 06/20", fontsize=20)
```

```python
daily_change = dataset.drop(['Sentiment', 'Direction', 'Signal'], axis=1)

daily_change.head()
```

```python
daily_change_plot = daily_change.plot(linewidth=1, figsize=(15,10));
daily_change_plot.set_xlabel("Date", fontsize=20)
daily_change_plot.set_ylabel("Change in closing price", fontsize=20)
daily_change_plot.set_title("Daily Change in Closing Price, 07/15 - 06/20", fontsize=20)
```

```python
signal = dataset.drop(['Sentiment', 'Daily Change', 'Direction'], axis=1)

signal.head()
```

```python
signal_plot = signal.plot(linewidth=1, figsize=(15,10));
signal_plot.set_xlabel("Date", fontsize=20)
signal_plot.set_ylabel("Signal", fontsize=20)
signal_plot.set_title("Daily Change in Direction, 07/15 - 06/20", fontsize=20)
```

```python
pos = 0
neg = 0
neutral = 0

for i in sentiment["Sentiment"]:
    if i > 0:
        pos += 1
    elif i <0:
        neg += 1
    else:
        neutral += 1
```

```python
plt.style.use('ggplot')

x = ['Positive', 'Neutral', 'Negative']
sentiment = [1133, 496, 178]

x_pos = [i for i, _ in enumerate(x)]

plt.bar(x_pos, sentiment, color='blue')
plt.xlabel("Sentiment Attribute")
plt.ylabel("Number of Days Featuring Sentiment Score")
plt.title("Sentiment Count per Day")

plt.xticks(x_pos, x)

plt.show()

plt.savefig('sentiment_count.pdf')  
```
